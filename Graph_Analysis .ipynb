```python


import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

df = pd.DataFrame({'Cars': [124,98,113]})
df.index = ['Mercedees','Audi','BMW']
df = df.sort_values(by='Cars',ascending=False)
df['freq']=df["Cars"]/df['Cars'].sum()*100
df["cumpercentage"] =df["freq"].cumsum().round()

fig, ax = plt.subplots(figsize=(8,7))
ax.bar(df.index, df["Cars"], color="black")
# plt.legend(['Mercedees','BMW',"Audi"],loc="upper left")

for v in  zip(df.index,df['Cars']):
    print(v,i)
    plt.annotate(f'{v[1]}',
    xy=(v[0],v[1]),
    color='black',
    fontsize=12             
#     ha="center",
#     va="bottom"             
    )
ax2 = ax.twinx()
ax2.plot( df["cumpercentage"], color="red", marker='o',label="commulative Line",linewidth=5)

for v in  zip(df.index,df['cumpercentage']):
#     print(v,i)
    plt.annotate(f'{v[1]}',
    xy=(v[0],v[1]),
    color='black',
    fontsize=12             
#     ha="center",
#     va="bottom"             
    )
plt.show()
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
```
